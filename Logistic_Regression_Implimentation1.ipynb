```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
print(data.DESCR)
```

```python
# Logistic regression till now we have studied with two values of the target variable
# This is called as Binary classification
```

```python
data.feature_names
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
df['target'] = data.target
```

```python
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample()
```

```python
df.target.unique()
```

```python
# for binary classification >> only two class needed, so we will filter out class 2
df = df[df['target'] != 2]
df
```

```python
df.target.unique()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
# train test split 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train
```

```python
# model training
from sklearn.linear_model import LogisticRegression
# open the documentation to learn about its parameters
```

```python
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
classifier.predict_proba(X_test) # corresponding to higher probability class is predicted, cuttoff is 0.5
# column 1 is prob of 0 and column 2 is prob of 1
```

```python
# Evaluation metrics
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# first two rows, 0 and 1 >> precision recall and f1 score and number of data points wrt respective class
# macro average >> for each class take simple avg >> avg precision of class 0 and class 1 will be macro avg
# simillarly for recall, f1 score
# support >> Total data points

# Weighted avg >> is used in case of class imbalance it gives importance to performance of dominant class
```

```python
from sklearn.metrics import roc_curve, auc

# get predicted probabilities of class 1 
classifier.predict_proba(X_test)[:,1] # probability of one class
```

```python
y_pred_proba = classifier.predict_proba(X_test)[:,1] # probability of one class
```

```python
y_pred_proba
```

```python
# reciever operating charecterstic curve 
# inputs : y_test, y_pred_proba. It will return TPR, FPR with different cutoff of probabilities


fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
```

```python
thresholds
```

```python
tpr
```

```python
fpr
```

```python
roc_auc = auc(fpr, tpr)
```

```python
plt.figure(figsize = (8, 6))
plt.plot(fpr, tpr, color = 'darkorange', linewidth = 2, label = 'ROC Curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color = 'navy', linewidth = 2, linestyle = '--')
plt.xlim([0.0, 1.0]) # Set x limits
plt.ylim([0.0, 1.05]) # Set y limits
plt.xlabel('False Positive rate')
plt.ylabel('True Positive Rate')
plt.title("ROC curve")
plt.legend(loc = 'lower right')
plt.show()
```
