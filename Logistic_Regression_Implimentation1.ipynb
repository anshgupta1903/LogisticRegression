```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
# Logistic regression till now we have studied with two values of the target variable
# This is called as Binary classification
```

```python
data.feature_names
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
df['target'] = data.target
```

```python
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample()
```

```python
df.target.unique()
```

```python
# for binary classification >> only two class needed, so we will filter out class 2
df = df[df['target'] != 2]
df
```

```python
df.target.unique()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
# train test split 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train
```

```python
# model training
from sklearn.linear_model import LogisticRegression
# open the documentation to learn about its parameters
```

```python
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
classifier.predict_proba(X_test) # corresponding to higher probability class is predicted, cuttoff is 0.5
# column 1 is prob of 0 and column 2 is prob of 1
```

```python
# Evaluation metrics
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
